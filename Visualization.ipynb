```python
# Sources/ References:
# For legend: 'Legend' section from https://github.com/matplotlib/AnatomyOfMatplotlib/blob/master/AnatomyOfMatplotlib-Part4-Limits_Legends_and_Layouts.ipynb
# For plotting x y coordinates: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
# For plotting polar coordinates: https://www.geeksforgeeks.org/plotting-polar-curves-in-python/#

import matplotlib.pyplot as plt
import numpy as np
```

```python
'''
Instructions:
The shape of a lima¸con can be defined parametrically as

  r = r0 + cos θ
  x = r cos θ
  y = r sin θ

Use this definition to plot the x-y shape of three lima¸cons, with

  r0 = 0.8,
  r0 = 1.0, and
  r0 = 1.2.

Be sure to use enough points that the curves are closed and appear smooth, except for the cusp.
Use a legend to identify which curve is which.

'''
```

# **Using x and y values:**

```python
def plotLimacon(r0_vals, colours):
  '''
  Function to plot the limacon for a given r0 value.

  Input:
    r0, [float]: r0 values to plot
    colours, [string]: list of colours to use for given r0 values

  Output:
    Plotting for give r0 values using x, y coordinates
  '''

  theta = np.linspace(0, 2*np.pi, 100) # Generate theta value for evaluation

  # Get radius value, and X Y axis data points
  for i,r0 in enumerate(r0_vals):
    r0_label = 'r0 = ' + str(r0)  # Generate labels
    c = colours[i]                # Get colours
    r = r0 + np.cos(theta)
    x = r * np.cos(theta)
    y = r * np.sin(theta)

    plt.plot(x, y, label=f'r0 = {r0}', color=c)

  plt.legend() # Add legend
  plt.axis('equal')  # Ensures x and y axes are scaled equally
  plt.savefig('limacon.png') # Save figure as png
  plt.show() # Show plot
```

```python
# Define the r0 values and their corresponding colours
r0_vals = [0.8, 1.0, 1.2]
colours = ['darkred', 'orange', 'pink']

plotLimacon(r0_vals, colours)
```

# **Using polar plotting**

```python

def plotLimaconPolar(r0_vals, colours):
  '''
  Function to plot the limacon for given r0 values using polar coordinates.

  Input:
    r0, [float]: r0 values to plot
    colours, [string]: list of colours to use for given r0 values

  Output:
    Polar plotting for give r0 value
  '''

  theta = np.linspace(0, 2*np.pi, 100) # Generate theta value for evaluation

  for i,r0 in enumerate(r0_vals):
    r0_label = 'r0 = ' + str(r0)  # Generate labels
    c = colours[i]                # Get colours
    r = r0 + np.cos(theta)
    x = r * np.cos(theta)
    y = r * np.sin(theta)

    # Use plt.polar to plot polar coordinates.
    plt.polar(theta, r, label=f'r0 = {r0}', color=c)

  plt.legend() # Add a legend
  plt.savefig('limacon_polar.png') #Save the plot as a png file

  plt.show() # Display the plot

```

```python
# Define the r0 values and their corresponding colours
r0_vals = [0.8, 1.0, 1.2]
colours = ['darkred', 'orange', 'pink']

plotLimaconPolar(r0_vals, colours)
```
