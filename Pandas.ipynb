# **Q1**

```python
# Resources: how to read in csv file in pandas: https://stackoverflow.com/questions/48340341/how-to-read-csv-to-dataframe-in-google-colab
url = "./College.csv"    # Read in College Data, assume it is in the files of the noteboook
data = pd.read_csv(url)
data    # Sanity check
```

```python
#(a) Which schools are in the top 5 for the percentage of accepted students who actually enroll?

top_id = data['Enroll'].div(data['Accept']).sort_values(ascending=False).nlargest(n=5).index # From pandas notebook
top_schools = data.iloc[top_id.values]['Unnamed: 0']

for x in top_schools.values:  # Show top schools
  print(x)
```

```python
#(b) Which are the top 5 colleges in terms of total cost (out-of-state tuition, room and board, books, and personal spending)?


top_expenses = (data['Outstate'] + data['Room.Board'] + data['Books'] + data['Personal']).sort_values(ascending=False)
top5 = top_expenses.nlargest(n=5).index
top_5_expenses = data.iloc[top5.values]['Unnamed: 0']

for x in top_5_expenses.values:  # Show schools with highest expenses
  print(x)
```

```python
#(c) Which are the top 5 colleges in terms of profit per student? This should be based only on out-of-state tuition revenue. Note that profit is revenue - cost.

top_profit = (data['Outstate'] - top_expenses).sort_values(ascending=False).nlargest(n=5).index
top_5_top_profit = data.iloc[top_profit.values]['Unnamed: 0']

for x in top_5_top_profit.values:  # Show schools with highest expenses
  print(x)
```

# **Q1**

```python
# Create a new column for the credit utilization for each person (balance divided by limit)

credit_data['Credit Util'] = credit_data['Balance'].div(credit_data['Limit']) # From pandas notebook
credit_data
```

# **Q2**

```python
# Resources: From "Computations with Categoricals" section of pandas notebook

# What is the average income for each education level?

ed_values = credit_data['Education'].sort_values()        # Sort by education level for clarity
income = credit_data['Income']                            # Get the income data
grade_bins = pd.Series(ed_values, name='education level') # Create a new series to store values

avg_income = (pd.Series(income) # By income
           .groupby(grade_bins) # Categorize by income
           .agg(['mean'])       # Aggregate by mean
           .reset_index())

print("Average incomes for each education level is: \n")
avg_income
```
